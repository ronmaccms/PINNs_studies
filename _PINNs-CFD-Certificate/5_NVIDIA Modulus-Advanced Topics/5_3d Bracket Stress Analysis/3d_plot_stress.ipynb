```python
import pandas as pd
```

```python
df1 = pd.read_csv("bracket.csv")
df2 = pd.read_csv("support.csv")
```

```python
df1
```

```python
df2
```

```python
merged_df = pd.concat([df1, df2], ignore_index = True)
```

```python
merged_df
```

```python
merged_df.columns
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure(figsize = (12,12))
ax = fig.add_subplot(projection = '3d')

scatter = ax.scatter(merged_df['Points_0'],merged_df['Points_1'],merged_df['Points_2'], c = merged_df['w'], cmap = 'rainbow', marker='o', s =10)

ax.set_xlabel('x_axis')
ax.set_ylabel('y_axis')
ax.set_zlabel('zaxis')

fig.colorbar(scatter, ax=ax, label = 'w')
```
