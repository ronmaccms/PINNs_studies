```python
import numpy
from matplotlib import pyplot
```

```python
length = 10
k = 0.89
# bonudary condition
temp_left = 100
temp_right = 200
total_sim_time = 10
```

image.png

```python
dx = .1
x_vector = numpy.linspace(0, length, int(length/dx))
```

```python
x_vector
```

```python
dt = 0.0001
t_vector = numpy.linspace(0, total_sim_time, int(total_sim_time/dt))
```

```python
t_vector
```

```python
# initial condition
u = numpy.zeros([len(t_vector),len(x_vector)])
```

```python
u.shape
```

```python
u
```

```python
u[0,0] = 100
u
```

```python
u[:,0] = temp_left
u[:,-1] = temp_right
```

```python
u
```

```python
pyplot.plot(x_vector, u[0])
pyplot.ylabel("temp")
pyplot.xlabel("x")
pyplot.show()
```

image.png

```python
for t in range(1, len(t_vector)-1):
  for x in range(1, len(x_vector)-1):
    u[t + 1, x] = ((k * (dt / dx**2)) * (u[t, x + 1] - 2 * u[t, x] + u[t,x-1])) + u[t, x]
```

```python
u
```

```python
pyplot.plot(x_vector, u[500])
pyplot.ylabel("temp")
pyplot.xlabel("x")
pyplot.show()
```

```python
pyplot.plot(x_vector, u[5000])
pyplot.ylabel("temp")
pyplot.xlabel("x")
pyplot.show()
```
