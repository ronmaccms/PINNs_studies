```python
import torch
```

```python
x = torch.tensor(1.0)
y = torch.tensor(2.0)

w = torch.tensor(1.0, requires_grad=True)
```

```python
print(x)
print(y)
print(w)
```

image.png

```python
y_hat = w*x
s = y_hat-y
loss = (s)**2
```

```python
print(loss)
print(loss.item())
```

```python
loss.backward()
```

```python
print(w.grad)
print(w.grad.item())
```
